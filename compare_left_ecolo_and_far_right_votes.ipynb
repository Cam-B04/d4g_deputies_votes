```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import requests
```

```python
bomparddf = pd.read_csv('manuel bompard.csv')
bomparddf.head()
```

```python
regoldf = pd.read_csv('Sandra Regol.csv')
regoldf.head()
```

```python
bomparddf.describe()
```

```python
df_left_votes = bomparddf[["vote_id", "for_or_against"]]
df_left_votes.head()
```

```python
df_ecolo_votes = regoldf[["vote_id", "for_or_against"]]
df_ecolo_votes.head()
```

```python
# Compare the votes of the two MEPs
df_votes = pd.merge(df_left_votes, df_ecolo_votes, on="vote_id", suffixes=('_bompard', '_regol'))
df_votes.head()
```

```python
# Find the votes where the two MEPs disagreed
df_disagree = df_votes[df_votes["for_or_against_bompard"] != df_votes["for_or_against_regol"]]
df_disagree.head(100)
```

```python
df_agree = df_votes[df_votes["for_or_against_bompard"] == df_votes["for_or_against_regol"]]
df_agree["for_or_against_bompard"].to_csv("agree_left_ecolo.csv")
# Rename the column for_or_against_bompard to left_ecolo_agreed_vote
df_agree.rename(columns={"for_or_against_bompard": "left_ecolo_agreed_vote"}, inplace=True)
df_agree["left_ecolo_agreed_vote"].head(100)
```
